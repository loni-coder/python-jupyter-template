### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x, y, color='blue')

# Annotate each point
for i in range(len(x)):
    plt.text(x[i], y[i], f'({x[i]}, {y[i]})', fontsize=9, ha='right')

# Add a title and labels
plt.title('Scatter Plot with Annotations')  # Plot title
plt.xlabel('X values')  # X-axis label
plt.ylabel('Y values')  # Y-axis label

# Show the plot
plt.show()
```
