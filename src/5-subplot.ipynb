### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 6)
y = np.array([2, 4, 6, 8, 10])
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line chart
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='b')
axs[0, 0].set_title("Line Chart")

# Bar chart
axs[0, 1].bar(categories, values, color='g')
axs[0, 1].set_title("Bar Chart")

# Scatter plot
axs[1, 0].scatter(x, y, color='r')
axs[1, 0].set_title("Scatter Plot")

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['blue', 'green', 'red', 'orange'])
axs[1, 1].set_title("Pie Chart")

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
